```python
import nltk;
from nltk import sent_tokenize
from nltk import word_tokenize
```

```python
file_path ="text.txt"
```

```python
with open(file_path, 'r', encoding='utf-8') as file:
    text = file.read()
```

```python
token = word_tokenize(text)
```

```python
token
```

```python
sentences = sent_tokenize(text)
```

```python
sentences
```

```python
#freq. of words
```

```python
from nltk.probability import FreqDist
tokens1 = word_tokenize(text)
freqOfWords = FreqDist(tokens1)
freqOfWords 
```

```python
import matplotlib.pyplot as plt
freqOfWords.plot()
```

```python
words_no_punc = []
for i in text:
    if i.isalpha():
        words_no_punc.append(i.lower())

print(words_no_punc)
```

```python
"after removing punctuation freq of words in text"
```

```python
freqOfWords.plot()
```

```python
from nltk.corpus import stopwords
stopwords = stopwords.words("german") # u can change here grman,english or any other lang...
stopwords
```

```python
clean_words = []
for i in words_no_punc:
    if i not in stopwords:
        clean_words.append(i)

clean_words
```

```python
from nltk.stem import PorterStemmer


porter = PorterStemmer()
word_list = ["Study", "Studying", "Studies"]
for word in word_list:
    print(f"{word} : {porter.stem(word)}")
```
